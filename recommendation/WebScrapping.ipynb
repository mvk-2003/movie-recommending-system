```python
import requests
from bs4 import BeautifulSoup
import csv
def fetch(url):
    src = requests.get(url)
    data = BeautifulSoup(src.text,'html.parser')
    return data

def pure(url):
    data = fetch(url)
    data = data.find('div',class_='site-body').find('div',class_='content-wrap').find('section').find('div',id="films-browser-list-container")
    data ='https://letterboxd.com'+data['data-url']
    return data

def extract(url):
    try:
        data = fetch(url)
        data = data.find('div',id='content').find('div',class_='content-wrap').find('div',class_='cols-3 overflow').find('div',class_='col-17')
        title = data.find('section','film-header-lockup -default').h1.text
        year = data.find('section','film-header-lockup -default').p.small.text
        director = data.find('section','film-header-lockup -default').p.span
        if(director!=None):
            director = director.text
        synopsis =data.find('section',class_='section col-10 col-main').find('div','review body-text -prose -hero prettify')
        if(synopsis!=None):
            synopsis = synopsis.find('div',class_='truncate').p.text
        cast,others = [],[]
        actors = data.find('section',class_='section col-10 col-main').find('div',id='tabbed-content').find('div',id='tab-cast')
        if(actors!=None):
            actors = actors.p.find_all('a',class_='text-slug tooltip')
            for i in actors:
                i = i.text
                cast.append(i)
        details = data.find('section',class_='section col-10 col-main').find('div',id='tabbed-content').find('div',id='tab-details').find_all('div',class_='text-sluglist')
        for i in details:
            i = i.p.find_all('a')
            others.append(i)
        if(len(others)==4):
            d1,d2,d3,d4 = others[0],others[1],others[2],others[3]
            studios,country,origin,language = [],[],[],[]
            for i in d1:
                i = i.text
                studios.append(i)
            for i in d2:
                i = i.text
                country.append(i)
            for i in d3:
                i = i.text
                origin.append(i)
            for i in d4:
                i = i.text
                language.append(i)
        elif(len(others)==3):
            d1,d2,d3 = others[0],others[1],others[2]
            studios,country,origin,language = [],[],[],[]
            for i in d1:
                i = i.text
                studios.append(i)
            for i in d2:
                i = i.text
                country.append(i)
            for i in d3:
                i = i.text
                language.append(i)
        elif(len(others)==2):
            d1,d2 = others[0],others[1]
            studios,country,origin,language = [],[],[],[]
            for i in d1:
                i = i.text
                country.append(i)
            for i in d2:
                i = i.text
                language.append(i)
        else:
            d1 = others[0]
            studios,origin,language = [],[],[]
            for i in d1:
                i = i.text
                language.append(i)
        genres,themes,temp = [],[],[]
        duplicate = data.find('section',class_='section col-10 col-main').find('div',id='tabbed-content').find('div',id='tab-genres')
        if(duplicate!=None):
            duplicate = duplicate.find_all('div','text-sluglist capitalize')
            for i in duplicate:
                i = i.p.find_all('a')
                temp.append(i)
            if(len(temp)==2):
                t1,t2 = temp[0],temp[1]
                for i in t1:
                    i = i.text
                    genres.append(i)
                for i in t2:
                    i = i.text
                    themes.append(i)
                themes.remove(themes[len(themes)-1])
            else:
                t1 = temp[0]
                for i in t1:
                    i = i.text
                    genres.append(i)
        final = {}
        final = {'title':[title],'year':[year],'director':[director],'synopsis':[synopsis],'cast':[cast],'studios':[studios],'country':[country],'origin':[origin],'language':[language],'genres':[genres],'themes':[themes]}
        return final
    except:
        return None

def load(d):
    csv_file = 'Telugu.csv'
    try:
        fields = ['title', 'year', 'director', 'synopsis','cast','studios','country','origin','language','genre','theme'] 
        with open(csv_file, 'a',newline='',encoding='utf-8') as csvfile: 
            writer = csv.writer(csvfile,delimiter='\t')  
            writer.writerows(zip(*d.values())) 
    except:
        return

def content(data):
    ajax_data = fetch(data)
    ajax_data = ajax_data.ul.find_all('li')
    for link in ajax_data:
        link ='https://letterboxd.com'+link.find('div')['data-film-slug']
        print(link)
        d = extract(link)
        load(d)

def travel(data):
    next_data = fetch(data)
    next_data = next_data.find('div',class_='pagination').find('a',class_='next')
    next_data ='https://letterboxd.com'+next_data['href']
    return next_data

data = pure('https://letterboxd.com/films/popular/language/telugu/')
while(data != None):
    content(data)
    data = pure(travel(data))
```
